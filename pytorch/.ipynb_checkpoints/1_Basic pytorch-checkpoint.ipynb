```python
torch.manual_seed(1)
r = torch.rand(2,2)
print(r)
r = r - 0.5 * 2
print(r)

print(torch.abs(r))

print(torch.asin(r))

print(torch.det(r))
print(torch.svd(r))
print(torch.max(r))
print(torch.std_mean(r))
```

```python
if torch.cuda.is_available():
    device = torch.device("cuda")
    x= torch.ones(5, device=device)
    print(x)
    y = torch.ones(5)
    y = y.to(device)
    z = x + y
    #if we want to convert the z tensor to numpy we first need
    #to bring it to cpu and then convert it to numpy
    z = z.to("cpu")
    z = z.numpy()
    z = z.to(device)
    
```
