```python
import torch
import torch.nn as nn
import numpy as np
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
data = datasets.load_breast_cancer()
X, y = data.data, data.target
print(y.shape)
n_samples, n_features = X.shape
print(n_samples, n_features)
X_train,X_test, y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=41)
print(X_train.shape)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
X_train = torch.from_numpy(X_train.astype(np.float32))
X_test = torch.from_numpy(X_test.astype(np.float32))
y_train = torch.from_numpy(y_train.astype(np.float32))
y_test = torch.from_numpy(y_test.astype(np.float32))
y_train=y_train.view(y_train.shape[0],1)
y_test=y_test.view(y_test.shape[0],1)
n_samples_train, n_features_train = X_train.shape
n_samples_test, n_features_test = X_test.shape
n_output_dim = y_train.shape[1]
```

```python
class LogisticRegression(nn.Module):
    def __init__(self,input_dim, output_dim):
        super(LogisticRegression,self).__init__()
        self.layer = nn.Linear(input_dim,output_dim)
    def forward(self,x):
        y_predicted = torch.sigmoid(self.layer(x))
        return y_predicted
```

```python
model = LogisticRegression(input_dim=n_features,output_dim = n_output_dim)
learning_rate = 0.03
loss_criterion = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(),lr=learning_rate)
```

```python
n_iters = 1000
loss_vals = []
for epoch in range(n_iters):
    y_pred = model(X_train)
    loss = loss_criterion(y_pred,y_train)
    loss.backward()
    loss_vals.append(loss.item())
    optimizer.step()
    optimizer.zero_grad()
    if (epoch+1) % 100 == 0:
        print(f'epoch : {epoch+1}, loss : {loss.item():.4f}')
plt.plot(range(n_iters),loss_vals)
plt.show()
```

```python
with torch.no_grad():
    y_predicted = model(X_test)
    y_predicted_cls = y_predicted.round()
    acc = y_predicted_cls.eq(y_test).sum() / float(y_test.shape[0])
    print(f'acuracy = {acc:.4f}')
```
