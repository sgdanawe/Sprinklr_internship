```python
import torch
```

```python
torch.manual_seed(10)
x = torch.randn(3, requires_grad=True)
print(type(x))
y = x + 2
print(y)
z = y*y*2
z = z.sum()
print(z)
z.backward() #dz/dx
print(x.grad)
zz = y*y*2
print(zz)
v = torch.ones(3)
# print(zz.shape)
# print(v.shape)
zz.backward(v)
print(x.grad)
```

```python
#removing the grad
#1
x.requires_grad_(False)
```

```python
weights = torch.ones(4, requires_grad=True)
for epoch in range(4):
    model_output = (weights*3).sum()
    model_output.backward()
    print(weights.grad)
    weights.grad.zero_()# very imp
```
