```python
import torch
```

```python
# we create a simple network for linear regression
# ground truth of the output is set to 2
#input is given as 1
# we need to find the weight so as to get the output 2 with x=1
# we first initialize the weight to 1
# we do a forward pass that is we compute the output using the hypothsis
# then we find the loss
# then we do a backward pass and update the weights
```

```python
x = torch.tensor(1.0)
y = torch.tensor(2.0)

w = torch.tensor(1.0, requires_grad=True)
```

```python
y_hat = w*x
loss = (y_hat-y)**2
print(loss)
```

```python
loss.backward()
print(w.grad)
```
